```python
ts = dataset['dow_jones_0.csv']
```

```python
ts.plot()
```

```python
plot_acf(ts.diff().dropna())
```

```python
d, q = 1, 0
```

```python
plot_pacf(ts.diff().dropna())
```

```python
p = 0
```

```python
from statsmodels.tsa import arima_model
from plotting import plot_multiple_ts
import pandas as pd
import numpy as np
```

```python
model = arima_model.ARIMA(ts, order=(p, d, q))
```

```python
res = model.fit()
```

```python
steps = 30
```

```python
forecast = res.forecast(steps=steps)
```

```python
prediction_index = pd.date_range(start=ts.index[-1], periods=steps, freq='D') 
```

```python
plot_multiple_ts(ts, pd.Series(forecast[0], index=prediction_index))
```
